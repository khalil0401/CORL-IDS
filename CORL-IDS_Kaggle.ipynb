# CORL-IDS: Continual Open-Set Reinforcement Learning Intrusion Detection System
---
This notebook implements the full Continual Open-Set RL framework from the CORL-IDS paper.
It uses a Rarity-Aware Soft Actor-Critic (SAC) reinforcement learning agent combined with a Self-Attention LSTM encoder.
It handles extreme class imbalance and uses adaptive Mahalanobis Centroid and Softmax Confidence detectors for zero-day open-set attack discovery.
Finally, it features an Elastic Weight Consolidation (EWC) continual learning loop to learn new attacks without catastrophic forgetting.

## 3. Full Continual Learning Pipeline (Phase 1, 2, and 3)
This executes the master orchestration script which rigorously performs:
- **Phase 1:** Trains the base RL agent on known traffic and logs metrics.
- **Phase 2:** Streams the hidden test set, flags unknowns using adaptive bounds, and evaluates the Zero-Day Detection Rate (UADR).
- **Phase 3:** Performs DBSCAN clustering to discover new categories, injects them into the Replay Buffer, and performs online Fine-Tuning using EWC to prevent forgetting.
