# CORL-IDS: Continual Open-Set Reinforcement Learning Intrusion Detection System
---
This notebook implements the framework from the CORL-IDS paper.
It uses a Rarity-Aware Soft Actor-Critic (SAC) reinforcement learning agent combined with an LSTM encoder.
It is designed to handle extreme class imbalance without data resampling (e.g., SMOTE) and includes a prototype novelty detector for zero-day open-set attack discovery.

## 3. Training the CORL-IDS Agent
This step trains the LSTM encoder and SAC agent purely on **known categories**. Imbalance is handled by the rarity-aware reward, not by mutating the dataset. It runs the Supervised Contrastive Loss to physically separate attack embeddings in the latent space.
By default, `ransomware` and `ddos` are completely hidden from training.

````python
!python src/train.py --epochs 100 --batch-size 256
````

## 4. Evaluation and Zero-Day Attack Discovery
The evaluation runs a rigorous Two-Phase pipeline:
- **Phase 1:** Classifies known network traffic correctly while aggressively flagging uncertain embeddings using Prototypes and Entropy.
- **Phase 2:** Runs density-based continual cluster discovery on the unknown flagged samples to identify the true hidden zero-day classes (e.g., `ransomware`, `ddos`).
